1. What is Simple Linear Regression (SLR)? Explain its purpose.
- Simple Linear Regression (SLR) is a statistical technique used to model the relationship between two variables by fitting a straight line to observed data.

Purpose of Simple Linear Regression

- Prediction

To predict the value of Y for a given X

Example: Predicting house price based on area

- Understanding Relationships

To measure the direction and strength of the relationship between X and Y

- Trend Analysis

To identify linear trends in data

- Decision Making

Helps businesses and researchers make data-driven decisions

6. What is Logistic Regression? How does it differ from Linear Regression?
- Logistic Regression is a statistical method used to predict the probability of a binary outcome (yes/no, 0/1, true/false) based on one or more independent variables. Unlike linear regression, it models categorical outcomes instead of continuous ones.

| Aspect                 | Linear Regression                       | Logistic Regression                                 |
| ---------------------- | --------------------------------------- | --------------------------------------------------- |
| **Dependent Variable** | Continuous (e.g., price, height)        | Binary/Categorical (e.g., yes/no, 0/1)              |
| **Output**             | Direct value of Y                       | Probability of class membership (0â€“1)               |
| **Equation**           | (Y = \beta_0 + \beta_1 X + \varepsilon) | (p = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}})      |
| **Error Assumption**   | Residuals are normally distributed      | Follows binomial distribution                       |
| **Goal**               | Minimize sum of squared errors          | Maximize likelihood (Maximum Likelihood Estimation) |
| **Curve Shape**        | Straight line                           | S-shaped (sigmoid) curve                            |
